```python
min_balnace=1000
print("welcome to the bank")
a=int(input("enter the pin"))
pin=1234
while True:
    if pin==1234:
        print("enter 1 for withdraw and 2 for enquiry")
        option=int(input("enter the option"))
        if option==1:
            amnt=int(input("enter the amount to withdraw"))
            if amnt>min_balnace:
                print("your balance is low")
            else:
                print("withdraw successful")
        elif option==2:
            balance=int(input("enter the balance"))
            if balance<min_balnace:
                print("your balance is low")
            else:
                print("your balance is sufficient")
    else:
        print("invalid option")    
```

```python
import csv

result = {}
row_count = 0

with open("data.csv", newline="") as file:
    reader = csv.reader(file)

    for row in reader:
        
        if row_count == 0:
            headers = row
            for h in headers:
                result[h] = []
        else:
            
            result[headers[0]] = result[headers[0]] + [row[0]]
            result[headers[1]] = result[headers[1]] + [(row[1])]
            result[headers[2]] = result[headers[2]] + [(row[2])]

        row_count += 1

print(result)
```
